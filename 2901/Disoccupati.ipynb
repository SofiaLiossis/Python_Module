```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_name ="./unemployment.csv"
unempl = pd.read_csv(file_name)
#print(unempl)
```

```python
#6)Richiesta di lavoro per distretto:

total = unempl["Number"].sum()
unempl["Percentage"] = unempl["Number"] / total * 100
richiesta_lavoro_per_disretto = unempl.groupby("District Name")["Number"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=richiesta_lavoro_per_disretto.index, y=richiesta_lavoro_per_disretto.values)
plt.title("Richiesta lavoro per distretto")
plt.xticks(rotation=90)
plt.show()
```

```python
#7) Richiesta di lavoro per genere:
richiesta_genere= unempl.groupby('Gender')['Number'].sum().reset_index()

# Grafico a torta
plt.pie(richiesta_genere['Number'], labels=richiesta_genere['Gender'], autopct='%1.1f%%', startangle=90)
plt.title('Domanda di lavoro per genere')
plt.show()
```

```python
#Andamento dei disoccuoati per genere:

sns.lineplot(data=unempl,x='Year',y='Number',hue='Gender',linewidth=5)

plt.title("Andamento disoccupati")
plt.show()
```
