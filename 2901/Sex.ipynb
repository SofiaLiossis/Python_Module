```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_name ="./immigrants_emigrants_by_sex.csv"
imm_sex = pd.read_csv(file_name)
print(imm_sex)
```

```python
#5) Immigrati suddivisi per sesso:
imm_per_sex = imm_sex.groupby("Gender")["Immigrants"].sum().sort_index(ascending=True)
print(imm_per_sex)
imm_per_sex.plot(kind='pie', autopct='%1.2f%%', startangle=90, colors=['violet', 'blue'])
plt.title("Immigrati per genere")
plt.ylabel("")
plt.show()
```

```python
imm_per_sex = imm_sex[['District Name','Gender','Immigrants','Emigrants']].groupby("District Name").sum().sort_values(by='Immigrants',ascending=False)
#print(imm_per_sex)

imm = imm_per_sex.reset_index()
plt.title('Immigrati vs Emigrati per Distretto')
sns.set_color_codes("muted")
sns.barplot(data=imm,x='Immigrants',y='District Name',label='Immigrati',color='b')
sns.set_color_codes("pastel")
sns.barplot(data=imm,x='Emigrants',y='District Name',label='Emigranti',color='g')

```
