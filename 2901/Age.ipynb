```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

file_name ="./immigrants_emigrants_by_age.csv"
imm_age = pd.read_csv(file_name)
#print(imm_age)
imm_age['Age'] = imm_age['Age'].replace(["5-9"],"05-09") #Modifica del valore per ordinarlo in maniera corretta
```

```python
# 4) Fascia di età degli immigrati per i vari distretti:

imm_per_eta = imm_age.groupby("Age")["Immigrants"].sum().sort_index(ascending=True)
#print(imm_per_eta)
imm_per_eta.plot(kind= 'bar', color='blue')
plt.title("Totale Immigrazione per fascia d'eta")
plt.xlabel('Numero di Immigrati')
plt.ylabel("Fascia d'eta")
plt.show()
```
